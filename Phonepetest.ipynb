```python
import os
import pandas as pd
import json
import psycopg2

# Aggregated Insurance

agg_ins_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/aggregated/insurance/country/india/state/"
agg_ins = os.listdir(agg_ins_path)

agg_ins_table = {"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[] }

for agg_ins_state in agg_ins:
    agg_ins_cur_state = agg_ins_path + agg_ins_state + "/"
    agg_ins_year_list = os.listdir(agg_ins_cur_state)
    

    for agg_ins_year in agg_ins_year_list:
        agg_ins_cur_year = agg_ins_cur_state + agg_ins_year + "/"
        agg_ins_file_list = os.listdir(agg_ins_cur_year)


        for agg_ins_file in agg_ins_file_list:
            agg_ins_cur_file = agg_ins_cur_year + agg_ins_file
            agg_ins_data = open(agg_ins_cur_file, "r")


            A = json.load(agg_ins_data)
            
            for i in A["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                agg_ins_table["Transaction_type"].append(name)
                agg_ins_table["Transaction_count"].append(count)
                agg_ins_table["Transaction_amount"].append(amount)
                agg_ins_table["States"].append(agg_ins_state)
                agg_ins_table["Years"].append(agg_ins_year)
                agg_ins_table["Quarter"].append(int(agg_ins_file.strip(".json")))

Aggregated_Insurance = pd.DataFrame(agg_ins_table)

Aggregated_Insurance["States"] = Aggregated_Insurance["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Aggregated_Insurance["States"] = Aggregated_Insurance["States"].str.replace("-" , " ")
Aggregated_Insurance["States"] = Aggregated_Insurance["States"].str.title()
Aggregated_Insurance["States"] = Aggregated_Insurance["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")



# Aggregated Transaction

agg_trans_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/aggregated/transaction/country/india/state/"
agg_trans = os.listdir(agg_trans_path)


agg_trans_table = {"States":[] , "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}


for agg_trans_state in agg_trans:
    agg_trans_cur_state = agg_trans_path + agg_trans_state + "/"
    agg_trans_year_list = os.listdir(agg_trans_cur_state)

    for agg_trans_year in agg_trans_year_list:
        agg_trans_cur_year = agg_trans_cur_state + agg_trans_year + "/"
        agg_trans_file_list = os.listdir(agg_trans_cur_year)
        

        for agg_trans_file in agg_trans_file_list:
            agg_trans_cur_file = agg_trans_cur_year + agg_trans_file
            agg_trans_data = open(agg_trans_cur_file, "r")


            B = json.load(agg_trans_data)

            for i in B["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                agg_trans_table["Transaction_type"].append(name)
                agg_trans_table["Transaction_count"].append(count)
                agg_trans_table["Transaction_amount"].append(amount)
                agg_trans_table["States"].append(agg_trans_state)
                agg_trans_table["Years"].append(agg_trans_year)
                agg_trans_table["Quarter"].append(int(agg_trans_file.strip(".json")))
                
Aggregated_Transaction = pd.DataFrame(agg_trans_table)

Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.replace("-" , " ")
Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.title()
Aggregated_Transaction["States"] = Aggregated_Transaction["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")



# Aggregated Users

agg_user_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/aggregated/user/country/india/state/"
agg_user = os.listdir(agg_user_path)


agg_user_table = {"States":[] , "Years":[], "Quarter":[], "Brands":[], "Transaction_count":[], "Percentage":[]}


for agg_user_state in agg_user:
    agg_user_cur_state = agg_user_path + agg_user_state + "/"
    agg_user_year_list = os.listdir(agg_user_cur_state)

    for agg_user_year in agg_user_year_list:
        agg_user_cur_year = agg_user_cur_state + agg_user_year + "/"
        agg_user_file_list = os.listdir(agg_user_cur_year)
        

        for agg_user_file in agg_user_file_list:
            agg_user_cur_file = agg_user_cur_year + agg_user_file
            agg_user_data = open(agg_user_cur_file, "r")


            C = json.load(agg_user_data)
            
            try:
                for i in C["data"]["usersByDevice"]:
                    Brand = i["brand"]
                    Count = i["count"]
                    Percentage = i["percentage"]
                    agg_user_table["Brands"].append(Brand)
                    agg_user_table["Transaction_count"].append(Count)
                    agg_user_table["Percentage"].append(Percentage)
                    agg_user_table["States"].append(agg_user_state)
                    agg_user_table["Years"].append(agg_user_year)
                    agg_user_table["Quarter"].append(int(agg_user_file.strip(".json")))
            except:
                pass
                
Aggregated_User = pd.DataFrame(agg_user_table)

Aggregated_User["States"] = Aggregated_User["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Aggregated_User["States"] = Aggregated_User["States"].str.replace("-" , " ")
Aggregated_User["States"] = Aggregated_User["States"].str.title()
Aggregated_User["States"] = Aggregated_User["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")



# Map Insurance


map_ins_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/map/insurance/hover/country/india/state/"
map_ins = os.listdir(map_ins_path)

map_ins_table = {"States":[], "Years":[], "Quarter":[], "Districts":[], "Transaction_count":[], "Transaction_amount":[]}

for map_ins_state in map_ins:
    map_ins_cur_state = map_ins_path + map_ins_state + "/"
    map_ins_year_list = os.listdir(map_ins_cur_state)


    for map_ins_year in map_ins_year_list:
        map_ins_cur_year = map_ins_cur_state + map_ins_year + "/"
        map_ins_file_list = os.listdir(map_ins_cur_year)


        for map_ins_file in map_ins_file_list:
            map_ins_cur_file = map_ins_cur_year + map_ins_file
            map_ins_data = open(map_ins_cur_file, "r")

            D = json.load(map_ins_data)

            for i in D["data"]["hoverDataList"]:
                district = i["name"]
                count = i["metric"][0]["count"]
                amount = i["metric"][0]["amount"]
                map_ins_table["States"].append(map_ins_state)
                map_ins_table["Years"].append(map_ins_year)
                map_ins_table["Quarter"].append(int(map_ins_file.strip(".json")))
                map_ins_table["Districts"].append(district)
                map_ins_table["Transaction_count"].append(count)
                map_ins_table["Transaction_amount"].append(amount)

Map_Insurance = pd.DataFrame(map_ins_table)

Map_Insurance["States"] = Map_Insurance["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Map_Insurance["States"] = Map_Insurance["States"].str.replace("-" , " ")
Map_Insurance["States"] = Map_Insurance["States"].str.title()
Map_Insurance["States"] = Map_Insurance["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Map_Insurance["Districts"] = Map_Insurance["Districts"].str.removesuffix("district")
Map_Insurance["Districts"] = Map_Insurance["Districts"].str.title()


# Map Transaction

map_trans_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/map/transaction/hover/country/india/state/"
map_trans = os.listdir(map_trans_path)

map_trans_table = { "States":[], "Years":[], "Quarter":[], "Districts":[], "Transaction_count":[], "Transaction_amount":[] }

for map_trans_state in map_trans:
    map_trans_cur_state = map_trans_path + map_trans_state + "/"
    map_trans_year_list = os.listdir(map_trans_cur_state)


    for map_trans_year in map_trans_year_list:
        map_trans_cur_year = map_trans_cur_state + map_trans_year + "/"
        map_trans_file_list = os.listdir(map_trans_cur_year)


        for map_trans_file in map_trans_file_list:
            map_trans_cur_file = map_trans_cur_year + map_trans_file
            map_trans_data = open(map_trans_cur_file, "r")


            E = json.load(map_trans_data)

            for i in E["data"]["hoverDataList"]:
                map_trans_district = i["name"]
                map_trans_count = i["metric"][0]["count"]
                map_trans_amount = i["metric"][0]["amount"]
                map_trans_table["States"].append(map_trans_state)
                map_trans_table["Years"].append(map_trans_year)
                map_trans_table["Quarter"].append(int(map_trans_file.strip(".json")))
                map_trans_table["Districts"].append(map_trans_district)
                map_trans_table["Transaction_count"].append(map_trans_count)
                map_trans_table["Transaction_amount"].append(map_trans_amount)


Map_Transaction = pd.DataFrame(map_trans_table)

Map_Transaction["States"] = Map_Transaction["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Map_Transaction["States"] = Map_Transaction["States"].str.replace("-" , " ")
Map_Transaction["States"] = Map_Transaction["States"].str.title()
Map_Transaction["States"] = Map_Transaction["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Map_Transaction["Districts"] = Map_Transaction["Districts"].str.removesuffix("district")
Map_Transaction["Districts"] = Map_Transaction["Districts"].str.title()


# Map User

map_user_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/map/user/hover/country/india/state/"
map_user = os.listdir(map_user_path)

map_user_table = { "States":[], "Years":[], "Quarter":[], "Districts":[], "Registered_Users":[], "App_opens":[] }

for map_user_state in map_user:
    map_user_cur_state = map_user_path + map_user_state + "/"
    map_user_year_list = os.listdir(map_user_cur_state)


    for map_user_year in map_user_year_list:
        map_user_cur_year = map_user_cur_state + map_user_year + "/"
        map_user_file_list = os.listdir(map_user_cur_year)


        for map_user_file in map_user_file_list:
            map_user_cur_file = map_user_cur_year + map_user_file
            map_user_data = open(map_user_cur_file)

            F = json.load(map_user_data)

            for i in F["data"]["hoverData"].items():
                district = i[0]
                registeredUser = i[1]["registeredUsers"]
                appOpens = i[1]["appOpens"]
                map_user_table["Districts"].append(district)
                map_user_table["Registered_Users"].append(registeredUser)
                map_user_table["App_opens"].append(appOpens)
                map_user_table["States"].append(map_user_state)
                map_user_table["Years"].append(map_user_year)
                map_user_table["Quarter"].append(int(map_user_file.strip(".json")))

Map_User = pd.DataFrame(map_user_table)

Map_User["States"] = Map_User["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Map_User["States"] = Map_User["States"].str.replace("-" , " ")
Map_User["States"] = Map_User["States"].str.title()
Map_User["States"] = Map_User["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Map_User["Districts"] = Map_User["Districts"].str.removesuffix("district")
Map_User["Districts"] = Map_User["Districts"].str.title()



# Top Insurance

top_ins_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/top/insurance/country/india/state/"
top_ins = os.listdir(top_ins_path)

top_ins_table = { "States":[], "Years":[], "Quarter":[], "Pincodes":[], "Transaction_count":[], "Transaction_amount":[] }

for top_ins_state in top_ins:
    top_ins_cur_state = top_ins_path + top_ins_state + "/"
    top_ins_year_list = os.listdir(top_ins_cur_state)


    for top_ins_year in top_ins_year_list:
        top_ins_cur_year = top_ins_cur_state + top_ins_year + "/"
        top_ins_file_list = os.listdir(top_ins_cur_year)


        for top_ins_file in top_ins_file_list:
            top_ins_cur_file = top_ins_cur_year + top_ins_file
            top_ins_data = open(top_ins_cur_file,"r")

            M = json.load(top_ins_data)
            
            for i in M["data"]["pincodes"]:
                pincode = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]
                top_ins_table["Pincodes"].append(pincode)
                top_ins_table["Transaction_count"].append(count)
                top_ins_table["Transaction_amount"].append(amount)
                top_ins_table["States"].append(top_ins_state)
                top_ins_table["Years"].append(top_ins_year)
                top_ins_table["Quarter"].append(int(top_ins_file.strip(".json")))

Top_Insurance = pd.DataFrame(top_ins_table)

Top_Insurance["States"] = Top_Insurance["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Top_Insurance["States"] = Top_Insurance["States"].str.replace("-" , " ")
Top_Insurance["States"] = Top_Insurance["States"].str.title()
Top_Insurance["States"] = Top_Insurance["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Top_Insurance["Pincodes"] = Top_Insurance["Pincodes"].str.title()



# Top Transaction

top_trans_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/top/transaction/country/india/state/"
top_trans = os.listdir(top_trans_path)

top_trans_table = { "States":[], "Years":[], "Quarter":[], "Pincodes":[], "Transaction_count":[], "Transaction_amount":[] }

for top_trans_state in top_trans:
    top_trans_cur_state = top_trans_path + top_trans_state + "/"
    top_trans_year_list = os.listdir(top_trans_cur_state)


    for top_trans_year in top_trans_year_list:
        top_trans_cur_year = top_trans_cur_state + top_trans_year + "/"
        top_trans_file_list = os.listdir(top_trans_cur_year)


        for top_trans_file in top_trans_file_list:
            top_trans_cur_file = top_trans_cur_year + top_trans_file
            top_trans_data = open(top_trans_cur_file, "r")

            N = json.load(top_trans_data)

            for i in N["data"]["pincodes"]:
                pincode = i["entityName"]
                count = i["metric"]["count"]
                amount = i["metric"]["amount"]
                top_trans_table["States"].append(top_trans_state)
                top_trans_table["Years"].append(top_trans_year)
                top_trans_table["Quarter"].append(int(top_trans_file.strip(".json")))
                top_trans_table["Pincodes"].append(pincode)
                top_trans_table["Transaction_count"].append(count)
                top_trans_table["Transaction_amount"].append(amount)

Top_Transaction = pd.DataFrame(top_trans_table)

Top_Transaction["States"] = Top_Transaction["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Top_Transaction["States"] = Top_Transaction["States"].str.replace("-" , " ")
Top_Transaction["States"] = Top_Transaction["States"].str.title()
Top_Transaction["States"] = Top_Transaction["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Top_Transaction["Pincodes"] = Top_Transaction["Pincodes"].str.title()



# Top Users

top_user_path = "C:/Users/Theju/Desktop/Capstone Projects/Phone pe/pulse/data/top/user/country/india/state/"
top_user = os.listdir(top_user_path)

top_user_table = { "States":[], "Years":[], "Quarter":[] , "Pincodes":[], "Registered_Users":[] }

for top_user_state in top_user:
    top_user_cur_state = top_user_path + top_user_state + "/"
    top_user_year_list = os.listdir(top_user_cur_state)


    for top_user_year in top_user_year_list:
        top_user_cur_year = top_user_cur_state + top_user_year + "/"
        top_user_file_list = os.listdir(top_user_cur_year)


        for top_user_file in top_user_file_list:
            top_user_cur_file = top_user_cur_year + top_user_file
            top_user_data = open(top_user_cur_file, "r")

            O = json.load(top_user_data)

            for i in O["data"]["pincodes"]:
                pincode = i["name"]
                registeredUsers = i["registeredUsers"]
                top_user_table["States"].append(top_user_state)
                top_user_table["Years"].append(top_user_year)
                top_user_table["Quarter"].append(int(top_user_file.strip(".json")))
                top_user_table["Pincodes"].append(pincode)
                top_user_table["Registered_Users"].append(registeredUsers)
                

Top_User = pd.DataFrame(top_user_table)

Top_User["States"] = Top_User["States"].str.replace("andaman-&-nicobar-islands" , "Andaman & Nicobar")
Top_User["States"] = Top_User["States"].str.replace("-" , " ")
Top_User["States"] = Top_User["States"].str.title()
Top_User["States"] = Top_User["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu" , "Dadra and Nagar Haveli and Daman & Diu")
Top_User["Pincodes"] = Top_User["Pincodes"].str.title()
```
